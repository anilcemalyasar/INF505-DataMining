**Berechnung des Mittelwerts fr eine Spalte:**

```python
mean_x1 = data['x1'].mean()
print(mean_x1)
```

```python
std_x1 = data['x1'].std()
print(std_x1)
```

```python
mean_x2 = data['x2'].mean()
print(mean_x2)
```

**Feature Engineering**
 Berechnung des Z-Scores fr jeden Wert einer Spalte und
speichern der Ergebnisse in einer neuen Spalte:


```python
data['z_x1'] = (data['x1'] - mean_x1) / std_x1
```

Mit der describe-Methode erhlt man einige ntzliche
Kennzahlen zu den Spalten des DataFrame-Objekts:


```python
data.describe()
```

```python
df_cols = [col for col in data.columns if col not in ['z_x1', 'y']]
for col in df_cols:
  data["z_"+col] = (data[col] - data[col].mean()) / data[col].std()
```

# Multiple Lineare Regression
Fr die Regression stehen verschiedene Bibliotheken, wie z.B.
sklearn oder statsmodels, zur Verfgung .

 Bei der statsmodels-API werden die exogenen und die
endogene Variablen an die Methode OLS bergeben und das
Modell berechnet.

 Die Regressionsparameter und Statistiken knnen dann mit der
summary-Methode fr das Ergebnis ausgegeben werden.

```python
import statsmodels.api as sm
data_x = data[['x1', 'x2', 'x3', 'x4']]
data_y = data['y']
data_x = sm.add_constant(data_x)
model = sm.OLS(data_y, data_x).fit()
print(model.summary())
```
