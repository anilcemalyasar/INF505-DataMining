## Wie wird die Zielvariable Status aufbereitet?

```python
credit_df = pd.read_csv("../datasets/credit_record.csv")
credit_df.head()
```

```python
credit_df["MONTHS_BALANCE"].sort_values(ascending=True)
```

```python
# X: No loan for the month
credit_df = credit_df[credit_df["STATUS"] != 'X']

# MONTHS_BALANCE: The month of the extracted data is the starting point, backwards, 0 is the current month, -1 is the previous month, and so on
credit_df = credit_df[credit_df["MONTHS_BALANCE"] > -12]
```

```python
credit_df['STATUS'].replace(['C'],0, inplace=True)
credit_df['STATUS'].replace(['0'],1, inplace=True)
credit_df['STATUS'].replace(['1'],2, inplace=True)
credit_df['STATUS'].replace(['2'],3, inplace=True)
credit_df['STATUS'].replace(['3'],4, inplace=True)
credit_df['STATUS'].replace(['4'],5, inplace=True)
credit_df['STATUS'].replace(['5'],6, inplace=True)
credit_df.drop("MONTHS_BALANCE", axis=1, inplace=True)
```

```python
credit_df = credit_df.groupby('ID').agg(max).reset_index()
credit_df.drop_duplicates(inplace=True)
credit_df
```

```python
# merge
final_df = pd.merge(app_df, credit_df, on='ID', how='inner')
```

```python
final_df.sort_values(by="ID", ascending=True).tail()
```

```python
ml_df = final_df[[
#    'ID',
    'FLAG_OWN_CAR',
    'FLAG_OWN_REALTY',
    'CNT_CHILDREN',
    'AMT_INCOME_TOTAL',
    'AGE_YEARS',
    'YEARS_EMPLOYED' ,
    'FLAG_MOBIL',
    'FLAG_WORK_PHONE',
    'FLAG_PHONE',
    'FLAG_EMAIL',
    'CNT_FAM_MEMBERS',
#    'NAME_INCOME_TYPE_0',
    'NAME_INCOME_TYPE_1',
    'NAME_INCOME_TYPE_2',
    'NAME_INCOME_TYPE_3',
    'NAME_INCOME_TYPE_4',
#    'NAME_EDUCATION_TYPE_0',
    'NAME_EDUCATION_TYPE_1',
    'NAME_EDUCATION_TYPE_2',
    'NAME_EDUCATION_TYPE_3',
    'NAME_EDUCATION_TYPE_4',
#    'NAME_FAMILY_STATUS_0',
    'NAME_FAMILY_STATUS_1',
    'NAME_FAMILY_STATUS_2',
    'NAME_FAMILY_STATUS_3',
    'NAME_FAMILY_STATUS_4',
#    'NAME_HOUSING_TYPE_0',
    'NAME_HOUSING_TYPE_1',
    'NAME_HOUSING_TYPE_2',
    'NAME_HOUSING_TYPE_3',
    'NAME_HOUSING_TYPE_4',
    'NAME_HOUSING_TYPE_5',
    'STATUS']]
ml_df.drop_duplicates()
```

```python
X = ml_df.drop(["STATUS"], axis=1)
y = ml_df["STATUS"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=35)
```

## Klassifikation mit Decision Tree, Random Forest, k-Nearest-Neighbor, Naive Bayes und XGBoost

```python
def classify(classifier):
    print(classifier)
    fit = classifier.fit(X_train, y_train)
    
    prediction = fit.predict(X_test)
    
    print(' Model Accuracy : ', fit.score(X_test, y_test)*100, '%')
    print("\nConfusion matrix: ")
    print(confusion_matrix(y_test, prediction))
    
    print("\nClassification Report: ")
    print(classification_report(y_true=y_test, y_pred=prediction))
    print('----------------------------------------')
```

```python
Classifiers = [
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    KNeighborsClassifier(),
    GaussianNB(),
    XGBClassifier()]

for c in Classifiers:
    classify(c)
```
